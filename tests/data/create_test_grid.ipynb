```python
import pandas as pd
import numpy as np
import os
```

# generators

```python
generators_df = pd.read_csv("etrago_test_network_1/generators.csv")
generators_df
```

## p

```python
base = np.ones((3,generators_df.shape[0]))
scale = np.array([0.0,0.5,1.0])

for i in range(scale.shape[0]):
    base[i,:] = scale[i] * base[i,:]
    
base
```

```python
generators_df["name"]
```

```python
power_df = pd.DataFrame(base, columns=generators_df["name"].to_list())
power_df
```

```python
power_df.to_csv("etrago_test_network_1/generators-p.csv")
power_df.to_csv("etrago_test_network_1/generators-p_min_pu.csv")
```
